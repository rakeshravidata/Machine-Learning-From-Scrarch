```python
#reading dataset
#importing Dataset
import numpy as np
import pandas as pd
import random
from sklearn.datasets import load_iris
from scipy.spatial import distance
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['class'] = iris.target
df['class.names'] = [iris.target_names[i] for i in iris.target]
df.drop('class', axis=1, inplace=True)
df.head()
df = df.drop("class.names",1)
```

```python
import random
num_clust = 4
centroid = []
num_features = 4
for i in range(0, num_clust):
    cent =[]
    for column in df:
        cent.append(random.randint(round(df[column].values.min()),round(df[column].values.max())))
    centroid.append(cent)
centroid_variation = [1,2,3,4]
p = 0
while np.mean(centroid_variation) > 0.001:
    print("#iteration")
    print(p+1)
    new_centroid = []
    centroid_variation = []
    clusters = []
    for index,row in df.iterrows():
        coordinates = []
        distances = []
        for column in range(0, num_features):
            coordinates.append(df.iloc[index,column])
        #print(coordinates)
        #print(centroid)
        for k in centroid:
            distances.append(distance.euclidean(coordinates,k))
        #print(distances)
        df.loc[index,"cluster"] = distances.index(min(distances)) + 1
        #print(df.loc[index,"cluster"])
    print("cluster - levels")
    df.cluster = df.cluster.astype(int)
    print(df["cluster"].unique())
    if len(df["cluster"].unique()) < num_clust:
        df = df.drop("cluster",1)
        for i in range(0, num_clust):
            cent =[]
            for column in df:
                cent.append(random.randint(round(df[column].values.min()),round(df[column].values.max())))
            centroid.append(cent)
        continue
    for i in range(1,num_clust+1):
        print(i)
        l = df.loc[df["cluster"]== i]
        n = l.shape[0]
        print(l.shape)
        m = []
        for column in l:
            m.append(l[column].mean())
        #print(m)
        new_centroid.append(m[0:4])
    print(new_centroid)
    for i in range(0, num_clust):
        #print(new_centroid[i])
        #print(centroid[i])
        centroid_variation.append(distance.euclidean(new_centroid[i],centroid[i]))
    #print("centroid variation")
    print(centroid_variation)
    centroid = new_centroid
    p = p+1
 
```
